```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Given actual and predicted values

 

actual = [176.800003, 179.580002, 177.330002, 169.240005, 166.570007]
predicted = [172.9657379867246, 177.2718762602242, 177.0275772388331, 177.9875973133142, 174.6921042390377]

# Convert to numpy arrays for easier calculations
actual = np.array(actual)
predicted = np.array(predicted)

# Calculate evaluation metrics
mae = mean_absolute_error(actual, predicted)
mse = mean_squared_error(actual, predicted)
rmse = np.sqrt(mse)

# Print evaluation metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Plot the actual vs predicted stock prices
dates = ['Nov 18, 2024', 'Nov 19, 2024', 'Nov 20, 2024', 'Nov 21, 2024', 'Nov 22, 2024']

plt.figure(figsize=(10, 6))
plt.plot(dates, actual, label='Actual Stock Price', color='blue', marker='o')
plt.plot(dates, predicted, label='Predicted Stock Price', color='red', marker='x')
plt.title('Actual vs Predicted Stock Prices for GOOG (Nov 18, 2024 - Nov 22, 2024)')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
